The p-value helps us measure how unusual our observed results are if the null hypothesis is true. 
Let's explain this in a simple way:

- **Null Hypothesis (\( H_0 \))**: This is the assumption we start with, often saying that there is no
    effect or no difference. For example, \( H_0 \) might state that the population mean (\( \mu \)) is 
    equal to a certain value.
  
- **Test Statistic**: This is a number we calculate from our sample data to compare to what we'd expect 
    under the null hypothesis. It tells us how far our sample data is from what we'd expect if \( H_0 \) were true.

- **Sampling Distribution**: If we assume the null hypothesis is true, the test statistic has a certain 
    distribution (its sampling distribution). This shows us the range of values we’d expect the test statistic to take on if \( H_0 \) is correct.

- **P-value**: The p-value is the probability of getting a test statistic as extreme as (or more extreme than) 
    what we observed, assuming the null hypothesis is true. 

Now, why does a smaller p-value make the null hypothesis look more "ridiculous"? 

If we calculate a very small p-value, it means the observed data is very unlikely under the assumption that 
the null hypothesis is true. It’s like saying, “If \( H_0 \) were true, it would be really strange to see data
like this.” The smaller the p-value, the more evidence we have that the null hypothesis might not be true, 
because the actual result we got is so far from what we'd expect under \( H_0 \). Therefore, a smaller p-value 
makes the null hypothesis look less believable.

In short: A small p-value means our observed result is unusual if the null hypothesis were true, making it 
    seem more likely that the null hypothesis is incorrect.

### **Relationship Between This Experiment and Fisher’s Original Experiment**

- In the STA130 student experiment, we expand the sample to 80 students and simplify the setup by having each student taste just one cup of tea.
- Both experiments aim to test a hypothesis of random guessing versus a detectable difference in taste based on the sequence of preparation.

### **Statements of the Null Hypothesis and Alternative Hypothesis**

#### **Null Hypothesis (\( H_0 \)):**
The null hypothesis states that students are merely guessing when identifying whether milk or tea was poured first. The population parameter of interest is the probability that a student guesses correctly by chance, which is \( p = 0.5 \).

**Formal version:**
- \( H_0: p = 0.5 \), where \( p \) is the proportion of students who correctly identify the sequence purely by random guessing.

**Informal version:**
- The null hypothesis suggests that students cannot truly taste a difference between pouring sequences and are just as likely to guess correctly as incorrectly, with a 50/50 chance of getting it right.

#### **Alternative Hypothesis (\( H_A \)):**
The alternative hypothesis suggests that students are better than random chance at identifying whether the milk or tea was poured first.

**Formal version:**
- \( H_A: p > 0.5 \), where \( p \) is the proportion of students who can correctly identify the sequence.

**Informal version:**
- The alternative hypothesis claims that students are not just guessing, but have a better-than-random ability to identify the pouring sequence.

---

### **Quantitative Analysis**
We will use a hypothesis test for a population proportion to assess whether 49 out of 80 correct responses are significantly different from what we would expect under random guessing.

#### **Methodology:**

1. **Observed Test Statistic**:
   - Sample proportion: \( \hat{p} = \frac{49}{80} = 0.6125 \)

2. **Null Hypothesis Sampling Distribution**:
   Under the null hypothesis, the sampling distribution of the sample proportion \( \hat{p} \) is approximately normal (due to the Central Limit Theorem) with:
   - Mean: \( \mu_{\hat{p}} = p_0 = 0.5 \)
   - Standard Error: \( \text{SE}_{\hat{p}} = \sqrt{\frac{p_0(1 - p_0)}{n}} = \sqrt{\frac{0.5(0.5)}{80}} \approx 0.0559 \)

3. **Test Statistic (z-score)**:
   The z-score compares how far the observed sample proportion is from the null hypothesis proportion in terms of standard errors:
   \[
   z = \frac{\hat{p} - p_0}{\text{SE}_{\hat{p}}} = \frac{0.6125 - 0.5}{0.0559} \approx 2.01
   \]

4. **P-value**:
   The p-value corresponds to the probability of observing a z-score of 2.01 or greater under the null hypothesis. From standard normal tables, the p-value is approximately \( 0.0222 \).

#### **Interpretation**:
- A p-value of 0.0222 means there is about a 2.22% chance of observing a sample result as extreme as ours (or more extreme) if students were merely guessing. 
- Since the p-value is less than the common significance level of \( \alpha = 0.05 \), we reject the null hypothesis, suggesting that students may indeed have some ability to detect the correct pouring sequence.

```python
import numpy as np
from scipy import stats

# Given data
n = 80  # Sample size
x = 49  # Number of correct responses
p_hat = x / n  # Sample proportion

# Null hypothesis value
p_0 = 0.5

# Standard error calculation
se = np.sqrt(p_0 * (1 - p_0) / n)

# z-test statistic
z = (p_hat - p_0) / se

# P-value calculation (for one-tailed test)
p_value = 1 - stats.norm.cdf(z)

print(f"Test Statistic (z): {z:.2f}")
print(f"P-value: {p_value:.4f}")
```
