```python
from sklearn import datasets
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

# Setting up the Data Frame
iris = datasets.load_iris()

iris_df = pd.DataFrame(iris.data, columns=['Sepal_Length',
                      'Sepal_Width', 'Petal_Length', 'Petal_Width'])

iris_df['Target'] = iris.target

iris_df['Target'].replace([0], 'Iris_Setosa', inplace=True)
iris_df['Target'].replace([1], 'Iris_Vercicolor', inplace=True)
iris_df['Target'].replace([2], 'Iris_Virginica', inplace=True)

# Plotting the KDE Plot
sns.kdeplot(iris_df.loc[(iris_df['Target']=='Iris_Virginica'),
            'Sepal_Length'], color='b', shade=True, label='Iris_Virginica')

# Setting the X and Y Label
plt.xlabel('Sepal Length')
plt.ylabel('Probability Density')
```

```python
import plotly.express as px
import plotly.graph_objects as go
import seaborn as sns
import numpy as np

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Prepare figure with histograms for each species
fig = px.histogram(penguins, x="flipper_length_mm", color="species", 
                   marginal="box", nbins=30, opacity=0.7)

# List of species
species_list = penguins['species'].unique()

# Loop through each species and add lines/rectangles
for species in species_list:
    # Filter data for the species
    species_data = penguins[penguins['species'] == species]['flipper_length_mm'].dropna()
    
    # Calculate statistics
    mean_val = species_data.mean()
    median_val = species_data.median()
    std_val = species_data.std()
    iqr_val = np.percentile(species_data, 75) - np.percentile(species_data, 25)
    min_val = species_data.min()
    max_val = species_data.max()
    
    # Range and IQR boundaries
    lower_bound_iqr = np.percentile(species_data, 25)
    upper_bound_iqr = np.percentile(species_data, 75)
    lower_bound_std = mean_val - 2 * std_val
    upper_bound_std = mean_val + 2 * std_val
    
    # Add lines for mean and median
    fig.add_vline(x=mean_val, line_dash="dash", line_color="blue", annotation_text=f"{species} Mean", annotation_position="top left")
    fig.add_vline(x=median_val, line_dash="dot", line_color="green", annotation_text=f"{species} Median", annotation_position="top right")
    
    # Add rectangles for range, IQR, and std deviations
    fig.add_vrect(x0=min_val, x1=max_val, line_width=0, fillcolor="lightblue", opacity=0.2, annotation_text=f"{species} Range", annotation_position="top right")
    fig.add_vrect(x0=lower_bound_iqr, x1=upper_bound_iqr, line_width=0, fillcolor="lightgreen", opacity=0.4, annotation_text=f"{species} IQR", annotation_position="top right")
    fig.add_vrect(x0=lower_bound_std, x1=upper_bound_std, line_width=0, fillcolor="yellow", opacity=0.2, annotation_text=f"{species} Â±2 Std", annotation_position="top right")

# Show the figure
fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the penguins dataset
penguins = sns.load_dataset('penguins')

# Create a figure with 3 subplots (one for each species) in a row
species = penguins['species'].unique()
fig, axes = plt.subplots(1, 3, figsize=(18, 5), sharey=True)

# Loop over each species to create KDE plots
for i, sp in enumerate(species):
    # Subset data for the species
    data = penguins[penguins['species'] == sp]['flipper_length_mm'].dropna()

    # Calculate statistics
    mean = data.mean()
    median = data.median()
    std_dev = data.std()
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    min_val = data.min()
    max_val = data.max()

    # Create KDE plot
    sns.kdeplot(data, ax=axes[i], fill=True, label=f'{sp} KDE', color='skyblue')

    # Plot vertical lines for mean and median
    axes[i].axvline(mean, color='red', linestyle='--', label='Mean')
    axes[i].axvline(median, color='blue', linestyle='--', label='Median')

    # Add shaded areas for range, IQR, and Â±2 standard deviations
    axes[i].axvspan(min_val, max_val, color='lightgray', alpha=0.3, label='Range')
    axes[i].axvspan(q1, q3, color='yellow', alpha=0.4, label='IQR')
    axes[i].axvspan(mean-2*std_dev, mean+2*std_dev, color='green', alpha=0.2, label='Â±2 STD')

    # Add titles and labels
    axes[i].set_title(f'{sp} Flipper Length KDE')
    axes[i].set_xlabel('Flipper Length (mm)')
    axes[i].legend()

# Set common Y-axis label
fig.text(0.04, 0.5, 'Density', va='center', rotation='vertical')

# Show the plot
plt.tight_layout()
plt.show()
```
